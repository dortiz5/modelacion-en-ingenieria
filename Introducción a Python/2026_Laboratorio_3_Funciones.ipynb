# Funciones

Las funciones son procedimientos o subprogramas que se utilizan para encapsular bloques de código, esto con el fin de poder re-utilizarlo en otras ocasiones.

**Ejemplo:** Crear la función sumar que consiste en sumar los elementos de una lista.

```python
def sumar(lista):
  S = 0
  for num in lista:
    S = S + num
  return S
```

```python
L1 = [17, 5, 23, 8]
L2 = [45, 12, 23, 18, 91]
print('La suma de la Lista 1 es: ', sumar(L1))
print('La suma de la Lista 2 es: ', sumar(L2))
```

**Ejemplo:** Crear una función que dado un número indique si es par o impar

```python
def par_impar(numero):
  #print('--> numero = ', numero)
  if numero%2==0:
    #print('-->--> Ingreso al if:', numero%2)
    print('par')
  else:
    #print('-->--> Ingreso al else:', numero%2)
    print('impar')
```

```python
numero = int(input('Ingrese un numero:'))
par_impar(numero)
```

**Ejemplo:** Crear una función que calcule el factorial de un número:

$$factorial(n) = 1\cdot 2\cdot 3\cdot ...\cdot (n-1) \cdot n$$

```python
list(range(3,7))
```

```python
def factorial(n):
  print('--> n=',n)
  fact = 1
  print('--> fact=',fact)
  for x in range(1,n+1):
    print('-->--> x=',x)
    fact = fact * x
    print('-->--> fact=',fact)
  print('-->Retorna fact=',fact)
  return fact
```

```python
numero = int(input('Ingrese un numero:'))
factorial(numero)
```

**Ejemplo:** Crean una función que calcula la combinatoria:

$$C^n_k = \frac{n!}{k!(n-k)!}$$

```python
def combinatoria(n,k):
  C = factorial(n)/(factorial(k)*factorial(n-k))
  return C
```

```python
n = int(input('Ingrese un numero:'))
k = int(input('Ingrese un numero:'))
print(combinatoria(n,k))
```

**Ejemplo:** Crear una función que calcule el promedio de los datos a partir de una lista, y luego otra función que calcule la varianza de los datos.

```python
def promedio(datos):
  prom = 0
  for k in datos:
    prom = prom + k
  prom = prom / len(datos)
  return prom
```

```python
def varianza(datos):
  var = 0
  for k in datos:
    var = var + k**2
  var = var /len(datos) - promedio(datos)**2
  return var
```

```python
import numpy as np

def var(datos):
  datos = np.array(datos)
  var = promedio((datos - promedio(datos))**2)
  return var
```

```python
D = [20, 15, 21, 32, 17, 8, 7, 13, 19, 24]
print('Promedio: ', promedio(D))
print('Varianza: ', varianza(D))
print('Varianza: ', var(D))
```

**Ejemplo:** Crear una función que encuentre el valor máximo de una lista

```python
def maximo(D):
  max = D[0]
  for k in D:
    if max<k:
      max = k
  return max
```

```python
maximo(D)
```
