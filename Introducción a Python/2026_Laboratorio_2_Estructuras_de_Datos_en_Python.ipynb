# 4. Tuplas <a class="anchor" id="tuplas"></a>

Una tupla es una secuencia de valores agrupados. Una tupla sirve para agrupar, como si fueran un único valor, varios valores que, por su naturaleza, deben ir juntos.

El tipo de datos que representa a las tuplas se llama tuple. El tipo tuple
es inmutable: una tupla no puede ser modificada una vez que ha sido creada.

```python
personaje = ('Mickey', 'Mouse')
personaje
```

Recuperación de los valores individuales

```python
nombre, especie = personaje
print('El nombre es ', nombre)
print('La especie es ', especie)
```

No se pueden cambiar los valores de la tupla, debido a que es inmutable

```python
personaje[1] = 'Cat' # El código lanzará un error
personaje
```

Los elementos de una tupla son ordenados según la ubicación de los elementos, para ello se usa el orden lexicográfico.

```python
elemento_1 = (3, -1, 7)
elemento_2 = (6, 2, 4)
elemento_3 = (3, -4, 9)
elemento_4 = (3, -1, 9)

print(elemento_1<elemento_2)
print(elemento_2<elemento_3)
print(elemento_1<elemento_3)
print(elemento_1<elemento_4)
```

Ejemplos de aplicaciones

```python
fecha = (2024, 'Marzo', 25)
print(fecha)
```

```python
import numpy as np

coordenada_1 = (0.78, -0.33)
coordenada_2 = (0.15, 1.8)

x1, x2 = coordenada_1
y1, y2 = coordenada_2

distancia = np.sqrt((x1-y1)**2+(x2-y2)**2)
print('Distancia = ', distancia)
```

# 5. Diccionarios <a class="anchor" id="diccionarios"></a>

Un diccionario puede ser visto como una colección de llaves, cada una
de las cuales tiene asociada un valor. Las llaves no están ordenadas y no hay
llaves repetidas. La única manera de acceder a un valor es a través de su llave.

```python
d1 = {
  "Nombre": "Sara",
  "Edad": 27,
  "Documento": 1003882
}
```

```python
d1["Nombre"]
```

```python
d1["Documento"]
```

Listar los elementos del Diccionario:

```python
for key in d1:
  print(key, d1[key])
```

```python
for key, valores in d1.items():
  print(key, valores)
```

Agregar un nuevo elemento al dicionario:

```python
d1['Direccion']='General Cruz 222'
d1
```

Obtener las llaves:

```python
d1.keys()
```

Obtener los valores

```python
d1.values()
```

```python
d1['Hermanos'] = ['Pedro', 'Juan']
d1
```

# 6. Nube de Palabras <a class="anchor" id="nubes"></a>

La nube de palabras es una presentación visual de las palabras más frecuentes de un documento.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud
```

```python
file = open('HarryPotter.txt','r')
libro = file.read()
file.close()
print(libro[:1000])
```

```python
word_cloud1 = WordCloud(collocations = False, background_color = 'white',
                        width = 2048, height = 1080).generate(libro)
# saving the image
word_cloud1.to_file('HarryPotter.png')
```

```python
plt.imshow(word_cloud1, interpolation='bilinear')
plt.axis("off")
plt.show()
```

# 7. Zipf Law <a class="anchor" id="zipf"></a>

La ley de Zipf es una ley empírica según la cual en una determinada lengua la frecuencia de aparición de distintas palabras sigue una distribución que puede aproximarse por

$$P_n \sim \frac{1}{n^a}$$


donde $P_n$ representa la frecuencia de la n-ésima palabra más frecuente y el exponente a es un número real positivo, en general ligeramente superior a 1.

Fue formulada en la década de 1940 por George Kingsley Zipf, lingüista de la Universidad de Harvard,

```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

def word_frequencies(words):

    return  Counter(words.split())


corpus = word_frequencies(libro)
Freq_words = corpus.values()
Freq = sorted(list(Freq_words), reverse = True)

N = 100
plt.bar(range(1,N+1),Freq[:N])

```
