* **Axes:** El sistema de coordenadas específico dentro de la figura.

### 2. Configuración Global de Estilo
El uso de `plt.rcParams` define un estándar visual académico en todo el notebook. Esto mejora la relación dato/tinta, elimina distracciones y emula la tipografía de LaTeX sin requerir una instalación nativa que pueda causar errores de compilación.




```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def configure_plot_style():
    gray = '#5c5c5c'
    plt.rcParams["mathtext.fontset"] = "cm"
    plt.rcParams["text.color"] = gray
    plt.rcParams["xtick.color"] = gray
    plt.rcParams["ytick.color"] = gray
    plt.rcParams["axes.labelcolor"] = gray
    plt.rcParams["axes.edgecolor"] = gray
    plt.rcParams["axes.spines.right"] = False
    plt.rcParams["axes.spines.top"] = False
    plt.rcParams['font.family'] = 'serif'
    plt.rcParams['mathtext.fontset'] = 'cm'
    plt.rcParams['font.size'] = 13

    plt.rcParams.update(
        {
            'text.usetex': False,
            'mathtext.fontset': 'stix',
        }
    )

configure_plot_style()
```

### 3. Gráficos de Línea (Plot Simple)
Para representar funciones continuas, generamos vectores espaciales con NumPy y los mapeamos a la figura.

```python

t = np.linspace(0, 1, 100)
f1 = 1
f2 = 2
y_sin = np.sin(2*np.pi*f1*t)
y_cos = np.cos(2*np.pi*f2*t)

plt.figure(figsize=(8, 4))
plt.plot(t, y_sin, label='Seno')
plt.plot(t, y_cos, label='Coseno', linestyle='--')

plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Comparación de Señales Trigonométricas')
plt.legend()

plt.tight_layout()
plt.savefig('solucion_temporal_pendulo.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

### 4. Subplots: Análisis Desacoplado
Cuando las señales tienen diferentes unidades o saturan el gráfico, `plt.subplots()` divide el lienzo en ejes independientes para un análisis más riguroso y ordenado.

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

# Subplot 1: Mezcla de línea continua y puntos de dispersión
ax1.plot(t, y_sin, color='gray', linestyle='--')
ax1.scatter(t, y_sin, color='C0', s=10)   # <-- s controla el tamaño
ax1.set_title('Componente Senoidal')
ax1.set_ylabel('Amplitud')
ax1.tick_params(axis='both', labelsize=14) 

# Subplot 2: Mezcla de línea continua y secuencia discreta (stem)
ax2.plot(t, y_cos, color='gray', alpha=0.5)
ax2.step(t,y_cos)          
ax2.set_title('Componente Cosenoidal')
ax2.set_xlabel('Tiempo [s]')
ax2.set_ylabel('Amplitud')
ax2.tick_params(axis='x', direction='in', length=8)

plt.tight_layout()
plt.show()
```

### 6. Campos vectoriales para retratos de fase

Un campo vectorial asigna un vector $\mathbf{v}(x, y)$ a cada punto del plano. En sistemas dinámicos, este vector corresponde a la derivada del estado — es decir, indica hacia dónde evoluciona el sistema desde cada condición inicial. Visualizarlos permite entender el comportamiento global sin necesidad de resolver las ecuaciones: se pueden identificar puntos de equilibrio, ciclos límite y regiones de atracción de un vistazo.

En Python, `plt.quiver` grafica los vectores $(U, V)$ sobre una grilla $(X, Y)$. Es común normalizar los vectores por su magnitud para que todos tengan el mismo largo y la dirección sea lo único visible — de lo contrario, los vectores grandes dominan la figura y ocultan la estructura local.

> Ejemplo adaptado del notebook [*2D Vector Visualization*](https://github.com/nicoguaro/taller_scivis-2025/blob/main/notebooks/03_2d_vector_visualization.ipynb) de Nicolás Guarín-Zapata.

```python
def ex_data(xmin=-2, xmax=2, ymin=-2, ymax=2, nx=21, ny=21):
    x = np.linspace(xmin, xmax, nx)
    y = np.linspace(ymin, ymax, nx)
    X, Y = np.meshgrid(x, y)
    U = Y
    V = -X - X**2
    return X, Y, U, V
```

```python
X, Y, U, V = ex_data()

mag = np.sqrt(U**2 + V**2)
U_norm = U/(mag + 1e-6) # Queremos evitar la división por cero
V_norm = V/(mag + 1e-6)
plt.figure(figsize=(6, 6))
plt.contourf(X, Y, mag, cmap="summer")
plt.quiver(X, Y, U_norm, V_norm)
plt.axis("image")
plt.xlabel("$x$")
plt.ylabel("$y$")
```

## Métodos de solución numéricos <a class="anchor" id="numerico"></a>

### ¿Por qué métodos numéricos?

La mayoría de los sistemas físicos reales son **no lineales**. Resolver sus ecuaciones analíticamente —si es que es posible— requiere matemática avanzada, soluciones implícitas y funciones especiales difíciles de evaluar. Los métodos numéricos nos permiten obtener trayectorias concretas sin necesidad de forma cerrada.

| | **Caso linealizado** | **Caso no lineal** |
|---|---|---|
| **Ecuación** | $\ddot{\theta} + \dfrac{b}{m}\dot{\theta} + \dfrac{g}{L}\theta = 0$, linealizando $\sin(\theta) \approx \theta$ | $\ddot{\theta} + \dfrac{b}{m}\dot{\theta} + \dfrac{g}{L}\sin(\theta) = 0$ |
| **Solución analítica** | $$\theta(t) = e^{-\gamma t}\left[A\cos(\omega_d t) + B\sin(\omega_d t)\right]$$ con $\gamma = \dfrac{b}{2m}$ y $\omega_d = \sqrt{\dfrac{g}{L} - \gamma^2}$ | $$t(\theta) = \sqrt{\frac{L}{g}}\, F(\varphi \mid k), \qquad F(\varphi \mid k) = \int_0^{\varphi} \frac{d\alpha}{\sqrt{1 - k^2\sin^2\!\alpha}}$$ con $k = \sin\!\left(\dfrac{\theta_0}{2}\right)$ y $\varphi = \arcsin\!\left(\dfrac{\sin(\theta/2)}{k}\right)$ |
| **Limitación** | Válida solo para $\theta_0$ pequeño. Falla para oscilaciones grandes o fricción no lineal. | $F(\varphi\mid k)$ es la **integral elíptica de primera especie** — función especial sin forma cerrada. Además, la solución entrega $t(\theta)$, no $\theta(t)$: hay que invertirla numéricamente de todas formas. |

### Métodos de solución numérica

La idea central proviene de la definición de derivada. Para $h$ pequeño pero finito, podemos **invertir** la definición y despejar el estado futuro:

$$\dot{\mathbf{x}}(t) = \lim_{h \to 0} \frac{\mathbf{x}(t+h) - \mathbf{x}(t)}{h} \quad \Longrightarrow \quad \mathbf{x}(t+h) \approx \mathbf{x}(t) + h\,\dot{\mathbf{x}}(t)\Longrightarrow\mathbf{x}(t+h) \approx \mathbf{x}(t) + h f(\mathbf{x}, t)$$

Más formalmente, cada método es una estrategia distinta para aproximar la integral subyacente:

$$\mathbf{x}(t_{n+1}) = \mathbf{x}(t_n) + \int_{t_n}^{t_{n+1}} f(\mathbf{x}(s), s)\, ds$$

Escribimos el problema en forma estándar: $\dot{\mathbf{x}} = f(\mathbf{x}, t)$, $\;\mathbf{x}(t_0) = \mathbf{x}_0$.


| | **[Euler explícito](https://en.wikipedia.org/wiki/Euler_method)** | **[Euler implícito]()** | **Runge-Kutta 4** |
|---|---|---|---|
| **Fórmula** | $\mathbf{x}_{n+1} = \mathbf{x}_n + h\,f(\mathbf{x}_n, t_n)$ | $\mathbf{x}_{n+1} = \mathbf{x}_n + h\,f(\mathbf{x}_{n+1}, t_{n+1})$ | $\mathbf{x}_{n+1} = \mathbf{x}_n + \dfrac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4)$ <br><br> $k_1 = f(\mathbf{x}_n, t_n)$ <br> $k_2 = f\!\left(\mathbf{x}_n + \tfrac{h}{2}k_1, t_n + \tfrac{h}{2}\right)$ <br> $k_3 = f\!\left(\mathbf{x}_n + \tfrac{h}{2}k_2, t_n + \tfrac{h}{2}\right)$ <br> $k_4 = f(\mathbf{x}_n + h\,k_3, t_n + h)$ |
| **Idea geométrica** | Rectángulo izquierdo: $f$ evaluada al inicio del intervalo | Rectángulo derecho: $f$ evaluada al final del intervalo; $\mathbf{x}_{n+1}$ desconocido, hay que resolver | Media ponderada de 4 pendientes dentro del intervalo (análogo a regla de Simpson) |
| **Error global** | $\mathcal{O}(h)$ | $\mathcal{O}(h)$ | $\mathcal{O}(h^4)$ |
| **Ventajas / Desventajas** | Trivial de implementar, 1 evaluación de $f$ / Acumula error rápido, puede diverger | Ideal para sistemas rígidos (*stiff*), permite pasos grandes / Requiere resolver un sistema en cada paso (`fsolve`) | Alta precisión con pasos moderados, estándar en la práctica / 4 evaluaciones de $f$ por paso |

Ahora, procederemos a programarlas. Para este caso, explicaremos únicamente el método Explícito y el método de Runge Kutta 

### Euler explícito — implementación

```python
def euler_explicito(f, x0, t):
    x = np.zeros((len(t), len(x0)))
    x[0] = x0
    for n in range(len(t) - 1):
        h = t[n+1] - t[n]
        x[n+1] = x[n] + h * f(x[n], t[n])
    return x
```

### Runge-Kutta 4 — implementación

```python
def runge_kutta4(f, x0, t):
    x = np.zeros((len(t), len(x0)))
    x[0] = x0
    for n in range(len(t) - 1):
        h = t[n+1] - t[n]
        k1 = f(x[n],t[n])
        k2 = f(x[n] + h/2 * k1, t[n] + h/2)
        k3 = f(x[n] + h/2 * k2, t[n] + h/2)
        k4 = f(x[n] + h * k3, t[n] + h)
        x[n+1] = x[n] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
    return x
```

Solución del péndulo con fricción lineal

$\ddot{\theta} + \dfrac{b}{m}\dot{\theta} + \dfrac{g}{L}\theta = 0$, linealizando $\sin(\theta) \approx \theta$

$\theta(t) = e^{-\gamma t}\left[A\cos(\omega_d t) + B\sin(\omega_d t)\right]$ con $\gamma = \dfrac{b}{2m}$ y $\omega_d = \sqrt{\dfrac{g}{L} - \gamma^2}$ 

Con $\theta(0)=\theta_0$ y $\dot{\theta}(0)=0$ tenemos $A = \theta_0, \quad B = \frac{\gamma\theta_0}{\omega_d}$

__Ejercicio:__ Implementar el código y comparar respecto a Euler implícito

```python
# ─── Parámetros del péndulo ───────────────────────────────────────────────────
g = 9.81   # gravedad [m/s²]
L = 1    # largo [m]
b = 1    # fricción [kg·m/s]
m = 1.0    # masa [kg]

# ─── Solución analítica ───────────────────────────────────────────────────────
def solucion_analitica(t, theta0, g, L, b, m):
    gamma  = b / (2*m)
    omegad = np.sqrt(g/L - gamma**2)
    A = theta0
    B = gamma * theta0 / omegad
    return np.exp(-gamma*t) * (A*np.cos(omegad*t) + B*np.sin(omegad*t))

# ─── Sistema linealizado: x = [θ, ω] ─────────────────────────────────────────
def f(x, t):
    theta, omega = x
    dtheta = omega
    domega = -(b/m)*omega - (g/L)*theta
    return np.array([dtheta, domega])

# ─── Simulación ──────────────────────────────────────────────────────────────
h_size = [0.1, 0.01]
theta0 = np.radians(15)  # condición inicial θ₀ = 15°
x0 = np.array([theta0, 0.0])
t_exacto = np.linspace(0, 15, 500)

fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
fig.suptitle(f"Péndulo linealizado con fricción")

for num, h in enumerate(h_size):
    t  = np.arange(0, 15, h)
    sol_euler = euler_explicito(f, x0, t)
    sol_rk4   = runge_kutta4(f, x0, t)
    analitica = solucion_analitica(t_exacto, theta0, g, L, b, m)

    # ─── Visualización ───────────────────────────────────────────────────────────
    axes[num].plot(t, sol_euler[:, 0], ".--", color="tomato", lw=1, label="Euler explícito")
    axes[num].plot(t, sol_rk4[:, 0], ".--", color="steelblue", lw=1, label="RK4")
    axes[num].plot(t_exacto, analitica, "k-", lw=2, label="Analítica")
    axes[num].set_title(f"Tamaño de h = {h}")
    axes[num].set_xlabel("Tiempo [s]")
    axes[num].set_ylabel("θ [rad]")
    axes[num].legend(loc = 'best')
    axes[num].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
